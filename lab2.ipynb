I have 1004991 rows and 28 columns. About half are numeric and the other half are nonnumeric. It has very precise geolocation data.

```python
#rename columns
df = df.rename(columns={"Crm Cd Desc": "Crime Description"})
df.head()
```

```python
df.isnull().sum() #show the number of rows containing null value
```

```python
df = df.dropna() #drop null values
df.isnull().sum()
```

I replaced the header "Crm Cd Desc" with "Crime Description". There was hundreds of thousands of missing values, so I removed them.

```python
df['Crime Description']
```

```python
df.sort_values(by='Crime Description') #sort data by a column
```

```python
df[0:5] #slice by rows
```

```python
df = df[(df['Crm Cd 1']>=0) & (df['Crm Cd 2']>=0) & (df['Crm Cd 3']>=0)] #filter data
df.describe()
```

I selected sorted and filtered by the different types of crimes. There are many more specifications of types of crimes than I previously thought.

```python
sum_time_by_crime = df.groupby('TIME OCC')['Crm Cd'].mean() # calculate the type of crime by time occured
sum_time_by_crime.head()
```

```python
ax = sum_time_by_crime.nlargest(10).plot.bar() # use a bar chart to show the results of the crimes
ax.set_xlabel("Time Occured")
ax.set_ylabel("Type of Crime")
ax.set_title("Crime Type by Time ")
```

This data shows time by crime type. The crimes that occured during the day were 210 (robbery). The crimes that occured late at night were mostly sexual crimes, indicated by their code numbers.

```python
sum_location_by_time = df.groupby('AREA')['TIME OCC'].mean() # calculate the crime occured by area
sum_location_by_time.head()
```

```python
ax = sum_location_by_time.nlargest(10).plot.bar() # use a bar chart to show the results of the areas and time of crimes committed
ax.set_xlabel("Area")
ax.set_ylabel("Time Committed")
ax.set_title("Area by Time Crime Committed")
```

This graph shows the average time that the crime was committed for each area code. Hollywood (6), for example shows that most crimes occure after 8:00 PM (2000 military time). Rampart (2), on the other hand shows most crimes occur midday.
